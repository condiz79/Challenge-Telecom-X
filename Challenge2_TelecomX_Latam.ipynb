```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import requests

# Cargar el dataset

url = "https://raw.githubusercontent.com/ingridcristh/challenge2-data-science-LATAM/main/TelecomX_Data.json"
response = requests.get(url)
data = response.json()
df = pd.DataFrame(data)

# Extract nested information into new columns
df['gender'] = df['customer'].apply(lambda x: x.get('gender'))
df['InternetService'] = df['internet'].apply(lambda x: x.get('InternetService'))
df['tenure'] = df['customer'].apply(lambda x: x.get('tenure'))
df['MonthlyCharges'] = df['account'].apply(lambda x: x.get('MonthlyCharges'))
df['TotalCharges'] = df['account'].apply(lambda x: x.get('TotalCharges'))

# Convert MonthlyCharges and TotalCharges to numeric, coercing errors
df['MonthlyCharges'] = pd.to_numeric(df['MonthlyCharges'], errors='coerce')
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce')


# Vista general
print("Primeras filas del dataset:")
display(df.head())
print("\nInformación general:")
print(df.info())
print("\nDistribución de Churn:")
print(df['Churn'].value_counts())

# Convertir columnas categóricas (excluyendo columnas con diccionarios)
categorical_cols = df.select_dtypes(include='object').columns
cols_to_convert = [col for col in categorical_cols if col not in ['customer', 'phone', 'internet', 'account']]
df[cols_to_convert] = df[cols_to_convert].astype('category')


# Estadísticas descriptivas
print("\nEstadísticas descriptivas:")
display(df.describe(include='all'))

# Visualizaciones
sns.set(style="whitegrid")

# Churn general
plt.figure(figsize=(6,4))
sns.countplot(data=df, x='Churn')
plt.title('Distribución de Churn')
plt.show()

# Churn por género
plt.figure(figsize=(6,4))
sns.countplot(data=df, x='gender', hue='Churn')
plt.title('Churn por Género')
plt.show()

# Churn por tipo de internet
plt.figure(figsize=(6,4))
sns.countplot(data=df, x='InternetService', hue='Churn')
plt.title('Churn por Tipo de Internet')
plt.show()

# Correlaciones numéricas
plt.figure(figsize=(10,8))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.title('Mapa de Calor de Correlaciones')
plt.show()

# Tenure vs Churn
plt.figure(figsize=(8,6))
sns.histplot(data=df, x='tenure', hue='Churn', bins=30, kde=True)
plt.title('Distribución de Tenure por Churn')
plt.show()

# Ingreso mensual estimado
if 'MonthlyCharges' in df.columns and 'TotalCharges' in df.columns:
    df['MonthlyIncome'] = df['TotalCharges'] / df['tenure'].replace(0, 1)

# Ingreso mensual vs Churn
if 'MonthlyIncome' in df.columns:
    # Filter out rows where 'Churn' is an empty string and where MonthlyIncome is NaN
    df_filtered = df[df['Churn'].str.strip() != ''].dropna(subset=['MonthlyIncome'])
    plt.figure(figsize=(8,6))
    sns.boxplot(data=df_filtered, x='Churn', y='MonthlyIncome')
    plt.title('Ingresos Mensuales por Churn')
    plt.show()

# Guardar dataset limpio
df.to_csv('TelecomX_Data_Clean.csv', index=False)
print("\nDataset limpio guardado como 'TelecomX_Data_Clean.csv'.")
```
